```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_error(map_name, condition_name, configs, errors):
    """
    Plots error curves for different experiment configurations with Seaborn style.

    Args:
        map_name (str): The name of the map.
        condition_name (str): The name of the experimental condition.
        configs (list): List of configuration dictionaries for the condition.
        errors (list of lists): List of error curves, each corresponding to a configuration.
    """
    # Set Seaborn style for the plots
    sns.set_theme()  
    
    # Check if the lengths of configs and errors match
    if len(configs) != len(errors):
        raise ValueError("The length of configs and errors must match.")
    
    # Prepare the data for plotting
    data = {
        'Iteration': [],
        'Error': [],
        'Configuration': []
    }
    
    for i, config_errors in enumerate(errors):
        config_label = f"Config {configs[i]}"  # Label for the configuration
        iterations = range(len(config_errors))  # Assuming error lists are based on iteration count
        data['Iteration'].extend(iterations)
        data['Error'].extend(config_errors)
        data['Configuration'].extend([config_label] * len(config_errors))
    
    # Convert the data into a Pandas DataFrame
    df = pd.DataFrame(data)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Iteration', y='Error', hue='Configuration', marker='o')

    # Customize the plot
    plt.title(f'Error Curves for {map_name} under {condition_name}')
    plt.xlabel('Iteration')
    plt.ylabel('Error')
    plt.legend(title='Experiment Configuration')
    
    # Show the plot
    plt.tight_layout()  # Adjust layout to not cut off labels
    plt.show()

# Example usage
# map_name = 'CityMap.png'
# condition_name = 'obs_size'
# configs = experiments['obs_size']
# errors = [[0.1, 0.08, 0.06], [0.15, 0.12, 0.10], [0.2, 0.18, 0.16]]  # Dummy error curves, replace with actual data
# plot_error(map_name, condition_name, configs, errors)
```

```python
for map in maps:
    for name, condition in experiments.items():
        condition_errors = []
        for config in condition:
            simulator = Simulator(map_img_file=map, obs_size=config['obs_size'], N=config['N'], dist_sim_weight=config['dist_sim_weight'])
            print('running exp with config: ', config)
            config_errors = simulator.simulate()
            condition_errors.append(config_errors)
            print(config_errors)
        plot_error(map_name=map, condition_name=name, configs=condition, errors=condition_errors)
```
